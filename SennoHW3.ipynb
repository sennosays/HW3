Interestingly, It takes more time to read the data in from memory than to write it to memory. Not surprisingly it takes less time to the data to a file than print it to the common output. The average time to read the file is reduced slightly compared to what it takes to read the file once. This may be that after the contents of the file have been loaded from memory onto the CPU it is faster to access the next time the file is to be read. 

```julia
data = randn(1024*1024); 
my_time_write_ASCII = time_file_write_ASCII(data,"ASCII_write_large"); 
```

```julia
my_time_read_ASCII = time_file_read_ASCII("ASCII_write_large"); 
```

The time required to write the file is increased by a factor of ~100 by doubling the size of the array, while the time required to read is increased by a factor of 755. Therefore, when we are scailing large problems the bottle neck will develop from reading the data in. This is fortunate because we normally only need to read data in once, while we may be writing results to the disk multiple times during a large calculation.

Comparing the benefits of writing data to an ASCII text file or a HDF5 jld file, it is obvious that the latter performs faster both in reading and writing, and also produces a significanly smaller file containing 1024^2 floating point numbers in an array. However, like in many cases the text file may take longer to write/read and take up more memory, but it can be read using a text editor. The jld file on the other hand cannot be opened except when it is being accessed by julia or some other HDF5 compatible software. Therefore, it would seem (like in many cases) that one should use an ASCII file to test and debug new code, and once scientific software passes the required tests and greater performance is really needed HDF5 should be used. HDF5 takes about 1/6th the amount of time to read the data in, and generates a file half the size of an ASCII text file.
